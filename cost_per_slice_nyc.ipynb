```python
from google.cloud import bigquery 
from google.colab import auth
auth.authenticate_user()
```

```python
project_id = 'play-368717'
dataset_id = 'mm_monday'
client = bigquery.Client(project=project_id)
```

```python
data_query = """
SELECT 
  Name AS name,
  CONCAT(location_lat,",",location_lng) AS location,
  Date_Expanded__times_in_EST_ AS day,
  Price AS price,
  Style AS type
FROM `play-368717.mm_monday.mm_pizza`
"""

df = client.query(data_query).to_dataframe()
```
